```python
import tensorflow as tf
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import glob
import random
print(tf.__version__)
```

```python
l2_reg = 0.0001
img_rows = 480
img_cols = 640
img_channels = 3
epochs = 3
batch_size = 8
```

```python
orig_image_path = 'orig_images'
hazy_image_path = 'hazy_images'

orig_image_paths = glob.glob(orig_image_path + '\\*.jpg')
n = len(orig_image_paths) 
print(f'Number of original images : {n}')
random.shuffle(orig_image_paths)
 
train_keys = orig_image_paths[:int(0.80*n)]
val_keys = orig_image_paths[int(0.80*n): int(0.90*n)]
test_keys = orig_image_paths[int(0.9*n):]

print(f'Number of original images for training : {len(train_keys)}')
print(f'Number of original images for validation : {len(val_keys)}')
print(f'Number of original images for testing : {len(test_keys)}')
```

```python
split_dict = {}
for key in train_keys:
    split_dict[key] = 'train'
for key in val_keys:
    split_dict[key] = 'val'
for key in test_keys:
    split_dict[key] = 'test'
```

```python
train_data = []
val_data = []
test_data = []
hazy_image_paths = glob.glob(hazy_image_path + '\\*.jpg')
for path in hazy_image_paths:
    label = path.split('\\')[-1]
    orig_path = orig_image_path + '\\' + label.split('_')[0] + '_' + label.split('_')[1] + ".jpg"
    if(split_dict[orig_path] == 'train'):
        train_data.append([path, orig_path])
    elif(split_dict[orig_path] == 'val'): 
        val_data.append([path, orig_path])
    else:
        test_data.append([path, orig_path])
```

```python
def load_image(x):
    x = tf.io.read_file(x)
    x = tf.image.decode_jpeg(x, channels=3)
    x = tf.image.resize(x, (480,640))
    x = x / 255.0
    return x

def show_image(x):
    x = np.asarray(x*255, dtype=np.int32)
    plt.figure()
    plt.imshow(x)
    plt.show()
```

```python
train_ds_hazy = tf.data.Dataset.from_tensor_slices([data[0] for data in train_data]).map(lambda x: load_image(x))
train_ds_orig = tf.data.Dataset.from_tensor_slices([data[1] for data in train_data]).map(lambda x: load_image(x))
train_ds = tf.data.Dataset.zip((train_ds_hazy,train_ds_orig)).shuffle(100).repeat().batch(batch_size)

val_ds_hazy = tf.data.Dataset.from_tensor_slices([data[0] for data in val_data]).map(lambda x: load_image(x))
val_ds_orig = tf.data.Dataset.from_tensor_slices([data[1] for data in val_data]).map(lambda x: load_image(x))
val_ds = tf.data.Dataset.zip((val_ds_hazy,val_ds_orig)).shuffle(100).repeat().batch(batch_size)
```

```python
iterator = tf.compat.v1.data.Iterator.from_structure(tf.compat.v1.data.get_output_types(train_ds),
                                           tf.compat.v1.data.get_output_shapes(train_ds))
train_init_op = iterator.make_initializer(train_ds)
val_init_op = iterator.make_initializer(val_ds)
next_element = iterator.get_next()
```

```python
inputs = tf.keras.layers.Input(shape=(img_rows, img_cols, img_channels))
conv1 = tf.keras.layers.Conv2D(3, kernel_size=1, activation='relu', padding='same', name='conv1',
                                kernel_initializer=tf.keras.initializers.RandomNormal(stddev=0.02),
                                kernel_regularizer=tf.keras.regularizers.l2(l2_reg))(inputs)
conv2 = tf.keras.layers.Conv2D(3, kernel_size=3, activation='relu', padding='same', name='conv2',
                                kernel_initializer=tf.keras.initializers.RandomNormal(stddev=0.02),
                                kernel_regularizer=tf.keras.regularizers.l2(l2_reg))(conv1)
concat1 = tf.keras.layers.concatenate([conv1, conv2], axis=-1, name='concat1')
conv3 = tf.keras.layers.Conv2D(3, kernel_size=5, activation='relu', padding='same', name='conv3',
                                kernel_initializer=tf.keras.initializers.RandomNormal(stddev=0.02),
                                kernel_regularizer=tf.keras.regularizers.l2(l2_reg))(concat1)
concat2 = tf.keras.layers.concatenate([conv2, conv3], axis=-1, name='concat2')
conv4 = tf.keras.layers.Conv2D(3, kernel_size=5, activation='relu', padding='same', name='conv4',
                                kernel_initializer=tf.keras.initializers.RandomNormal(stddev=0.02),
                                kernel_regularizer=tf.keras.regularizers.l2(l2_reg))(concat2)
concat3 = tf.keras.layers.concatenate([conv1, conv2, conv3, conv4], axis=-1, name='concat3')
k_x = tf.keras.layers.Conv2D(3, kernel_size=5, activation='relu', padding='same', name='k_x',
                                kernel_initializer=tf.keras.initializers.RandomNormal(stddev=0.02),
                                kernel_regularizer=tf.keras.regularizers.l2(l2_reg))(concat3)

inter_1 = tf.keras.layers.Subtract()([tf.keras.layers.Multiply()([k_x, inputs]),k_x])
outputs = tf.keras.layers.Lambda(lambda x: tf.keras.activations.relu(x))(tf.keras.layers.Lambda(lambda x: 1+x)(inter_1))
model = tf.keras.models.Model(inputs=inputs, outputs=outputs)
```

```python
model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
loss_function = tf.keras.losses.MeanSquaredError()

train_batches = len(train_data)//batch_size
validation_batches = len(val_data)//batch_size
print(f'Number of batches of training is : {train_batches}')
print(f'Number of batches of validation is : {validation_batches}\n')

for epoch in range(epochs):
    print(f'Starting Epoch Number: {epoch+1}\n')
    for i in range(train_batches):
        with tf.GradientTape() as tape:
            outputs = model(next_element[0], training=True)
            train_loss = loss_function(outputs, next_element[1])
        grads = tape.gradient(train_loss, model.trainable_weights)
        optimizer.apply_gradients(zip(grads, model.trainable_weights))
        if i % 100 == 0:
            print(f'Training Loss : {train_loss} at {i} iterations')
    print('\n')
    for j in range(validation_batches):
        outputs = model(next_element[0], training=False)
        validation_loss = loss_function(outputs, next_element[1])
        if j % 100 == 0:
            print(f'Validation Loss : {validation_loss} at {j} iterations')
    print('\n')  
```

```python
def load_image_test(x):
    x = tf.io.read_file(x)
    x = tf.image.decode_jpeg(x, channels=3)
    x = tf.image.resize(x, (480,640))
    return x

psnr = []
ssim = []
for (hazy, orig) in test_data:
    hazy = cv2.imread(hazy)
    hazy = hazy.reshape(1, hazy.shape[0], hazy.shape[1], hazy.shape[2])
    hazy = tf.image.resize(hazy, (480,640))/255.0
    dehazed = model(hazy)
    dehazed = dehazed*255
    dehazed = dehazed.numpy()
    dehazed = dehazed.reshape(dehazed.shape[1], dehazed.shape[2], dehazed.shape[3])
    dehazed = tf.convert_to_tensor(dehazed)
    orig = load_image_test(orig)
    psnr.append(tf.image.psnr(dehazed, orig, max_val=255))
    
    dehazed = dehazed.numpy()
    orig = orig.numpy()
    dehazed = cv2.cvtColor(dehazed, cv2.COLOR_RGB2GRAY)
    orig = cv2.cvtColor(orig, cv2.COLOR_RGB2GRAY)
    dehazed = dehazed.reshape(dehazed.shape[0], dehazed.shape[1], 1)
    orig = orig.reshape(orig.shape[0], orig.shape[1], 1)
    dehazed = tf.convert_to_tensor(dehazed)
    orig = tf.convert_to_tensor(orig)
    ssim.append(tf.image.ssim(dehazed, orig, max_val=255))

print(f'Average PSNR is : {sum(psnr)/len(psnr)}')
print(f'Average SSIM is : {sum(ssim)/len(ssim)}')
```
